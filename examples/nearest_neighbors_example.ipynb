```python
"""
Today we will be looking at K-Nearest-Neighbors, which is an extremely simple algorithm for classification (and regression).
All it does is just look at the k nearest points (you decide the k), and find the average of what class they are in or
their regression label. The way it finds the nearest points are through a euclidean distance formula. Another thing
to note is that the k value is usually set to be odd, as if it is even : what do you do if there is a tie? Other than 
that KNN is a walk in the park. 
"""

# let's get started
import sealion as sl 
from sealion.nearest_neighbors import KNearestNeighbors
```

```python
#we'll use the blobbed dataset
from sklearn.datasets import make_blobs
X, y = make_blobs(1000, centers = 10, random_state = 11)
```

```python
# next let's visualize our datasets
import matplotlib.pyplot as plt
fig = plt.figure() 
ax = fig.add_subplot()
ax.scatter(X[:, 0], X[:, 1])
plt.title("Blobs Dataset")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#we'll split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size = 0.2, random_state = 7)
```

```python
# next we can simply just apply the algorithm. 
knn = KNearestNeighbors(k = 5) # let's try k = 5
knn.fit(X_train, y_train)
```

```python
# let's evaluate it : 
print(knn.evaluate(X_test, y_test))
knn.visualize_evaluation(knn.predict(X_test), y_test)
```

```python
# why don't we see what happens when we use different k-values? 

k_values = [3, 4, 5, 7, 9, 11, 13]

for k_value in k_values : 
    knn = KNearestNeighbors(k = k_value)
    knn.fit(X_train, y_train)
    print(f"{k_value} : {knn.evaluate(X_test, y_test)}")
```

```python
# Looks like having a too high k-value makes it worse. That's probably because if you look at too many other points you 
# maybe looking at points in order clusters. 
# The best seems to be 3 and 4.  
```
